```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
""" Load the iris dataset using Seaborn's "load_dataset" function  """
iris = sns.load_dataset('iris')
#### Create input and output data arrays for the machine learning model
# Input features (sepal length, sepal width, petal length, petal width)
X_iris = iris.drop('species', axis = 1) 
# Target variable (iris species)
y_iris = iris['species'] 

### Define a set of random input and output data points for a scatter plot
rng = np.random.RandomState(35)
x = 10*rng.rand(40)
y = 2*x-1+rng.randn(40)
```

```python
""" Create a PCA object with 2 principal components """
model = PCA(n_components=2)

## Fit the PCA model to the input data
model.fit(X_iris)
# Apply the learned PCA transformation to the input data
X_2D = model.transform(X_iris)
## Add the transformed data to the iris dataset as new columns
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]

print(iris['PCA1'][:10])
print(iris['PCA2'][:10])
```

```python
""" Plot the transformed data with the iris species as the color hue.
N.B.
    - lmplot = Scatter plot data and regression model fits across a FacetGrid
    - General rule = when assign variables to different facets, is to use ``hue`` for the most important comparison, 
        followed by ``col`` and ``row``.
    - use the comma ; to avoid print <seaborn.axisgrid.FacetGrid at 0x7fbab1353d00> object
    - x_ci = Size of the confidence interval used when plotting a central tendency
"""
#sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, fit_reg=False); 
sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, height=6, aspect=1.2, ci=None, scatter_kws={"s": 80}, legend=False);
# Add a custom legend
plt.legend(title='Species', loc='upper right')
```

```python
""" Visualize PCA points.
N.B.
With "fit_reg=False" Seaborn will only display the scatter plot of the data points without fitting a regression line.
> Add legend automatically
"""
sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, fit_reg=False);
```

```python
""" Visualize PCA """
sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, height=6, aspect=1.2, ci=None, scatter_kws={"s": 80});
plt.plot([-4, 4], [0, 0], ':k')
plt.plot([0, 0], [-4, 4], ':k')
plt.xlim([-4, 4])
plt.ylim([-4, 4])

# Add a legend with a title and move it to the upper right corner
#plt.legend(title='Species', loc='upper right')
plt.show()
```

```python
# Load the iris dataset
iris = sns.load_dataset('iris')

##### Perform PCA on the iris dataset
model = PCA(n_components=2)
model.fit(iris.iloc[:, :-1])
X_2D = model.transform(iris.iloc[:, :-1])
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
```

```python
"""  Display the lmplot.
N.B.1
=> Add a legend with a title and move it to the upper right corner
#plt.legend(title='Species', loc='upper right')
N.B.2
=> Place legend on the right, but centred (bbox = (distance from x, distance from y)  
#plt.legend(title='Species', bbox_to_anchor=(0.87, 1), loc='upper right') 
"""
#### Create the scatter plot with principal components and regression lines
sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, height=6, aspect=1.2, ci=None, scatter_kws={"s": 80}, legend=False)
plt.plot([-4, 4], [0, 0], ':k')
plt.plot([0, 0], [-4, 4], ':k')
# Set the limits for x and y axes
plt.xlim([-4, 4])
plt.ylim([-4, 4])

# Place legend on the extreme right
plt.legend(title='Species', bbox_to_anchor=(0.87, 1), loc='upper left')  # 

# Show the plot #useless here in notebook
plt.show()


sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, fit_reg=False);
sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, height=6, aspect=1.2, ci=None, scatter_kws={"s": 80});
plt.plot([-4, 4], [0, 0], ':k')
plt.plot([0, 0], [-4, 4], ':k')
plt.xlim([-4, 4])
plt.ylim([-4, 4])

# Add a legend with a title and move it to the upper right corner
plt.legend(title='Species', loc='upper right')
plt.show()
```
