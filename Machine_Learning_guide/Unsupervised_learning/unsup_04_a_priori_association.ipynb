Apriori is an algorithm for frequent item set mining and association rule learning over transactional databases.      
It is used to find sets of items or itemsets that frequently occur together in transactional data. 

- min_support: Minimum support threshold for an itemset to be considered frequent
- min_confidence: Minimum confidence threshold for association rules
- min_lift: Minimum lift threshold for association rules
- min_length: Minimum number of items in an itemset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
dataset_book = pd.read_csv('./data_to_use/books_scraped.csv')
dataset_book.head()
```

```python
# Reduce the dataset for a quick trials
dataset_book = dataset_book[:20]
len(dataset_book)
```

```python
# Extract values from columns 0 to 4 
transactions=[]
for i in range(0, 20):
    transactions.append([str(dataset_book.values[i, j]) for j in range(0, 5)])
```

```python
transactions
```

```python
rule = apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)
```

```python
# Convert the discovered rules into a list
the_rules_are = list(rule) 
the_rules_are
```
