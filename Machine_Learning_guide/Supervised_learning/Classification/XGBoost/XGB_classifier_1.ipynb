```python
import pandas as pd
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score

import xgboost as xgb
```

```python
df = pd.read_csv("./wisc_bc_data.csv")
df.describe().T
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# Rename Dataset to label to make it easy to understand
df = df.rename(columns={'diagnosis':'label'})
print(df.dtypes)
```

```python
df.head()
```

```python
# Understand the data 
sns.countplot(x="label", data=df)
```

```python
""" Define X set and normalize values """
X = df.drop(labels = ["label", "id"], axis=1) 

scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
"""  Replace categorical values with numbers. """
df['label'].value_counts()

# Define the dependent variable that needs to be predicted (labels)
y = df["label"].values
## Encoding categorical data
labelencoder = LabelEncoder()
Y = labelencoder.fit_transform(y) 
```

```python
""" Split data into train and test. """
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
# Create a Random Forest model, to compare it against xgboost
model1 = RandomForestClassifier(n_estimators = 10, random_state = 42)
```

```python
# Train random forest model on training data
model1.fit(X_train, y_train);
```

```python
# Make predictions for random forest model on test data
y_pred1 = model1.predict(X_test)
```

```python
## Train XGboost
model = xgb.XGBClassifier()
model.fit(X_train, y_train)
```

```python
# Predicting XGboost on test set results
y_pred = model.predict(X_test)
```

```python
# Evaluate the classifier on test data RandomForest
accuracy = accuracy_score(y_test, y_pred1)
print("Accuracy = ", (accuracy * 100.0), "%")
```

```python
# Confusion Matrix RandomForest
cm = confusion_matrix(y_test, y_pred1)
my_cmap = 'Blues'
sns.heatmap(cm, annot=True, cmap=my_cmap);
```

```python
# Evaluate the classifier on test data XGboost
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy = ", (accuracy * 100.0), "%")
```

```python
# Confusion Matrix XGboost
cm = confusion_matrix(y_test, y_pred)
#my_cmap = 'Greens'
my_cmap = 'YlGnBu'
#my_cmap = 'Oranges'
#my_cmap = 'plasma'
sns.heatmap(cm, annot=True, cmap=my_cmap)
```
