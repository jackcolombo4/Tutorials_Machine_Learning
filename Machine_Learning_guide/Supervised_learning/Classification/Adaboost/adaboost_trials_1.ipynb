<h1 style="color:#C2F5A6  "> Adaboost trials 1 </h1>

```python
import pandas as pd
from copy import copy
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report

from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
df_cars = pd.read_csv('./car_evaluation.csv', header=None)
df_cars.head()
```

```python
headers_cars = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
df_cars.columns = headers_cars
df_cars1 = copy(df_cars)
df_cars.head()
```

#### => Encode category feautures

```python
mask = df_cars['persons'] == "more"
mask
```

```python
map1 = {'vhigh':3, 'high':2, 'med':1, 'low':0}
df_cars['buying'] = df_cars['buying'].map(map1)
df_cars['maint'] = df_cars['maint'].map(map1)
df_cars.head()
```

```python
le = LabelEncoder()
df_cars['lug_boot'] = le.fit_transform(df_cars['lug_boot'])
df_cars.head()
```

```python
ohe = OneHotEncoder(sparse_output=False)
df_cars['safety'] = ohe.fit_transform(df_cars[['safety']]) 
df_cars['persons'] = ohe.fit_transform(df_cars[['persons']]) 
df_cars['doors'] = ohe.fit_transform(df_cars[['doors']]) 
df_cars
```

```python
## Split into features and target
X = df_cars.iloc[:,:-1] 
y = df_cars['class']
# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y) 
```

```python
ada = AdaBoostClassifier(n_estimators=100, learning_rate=1)

# Train the model 
ada.fit(X_train, y_train)
# Make predictions
y_pred = ada.predict(X_test)
```

```python
""" Metrics """
# Model Accuracy
print('Accuracy: ', ada.score(X_test, y_test))

# Detailed classification report
print(classification_report(y_test, y_pred))

# Feature importance
print('Feature Importance: \n', ada.feature_importances_)
```

```python
le = LabelEncoder()
df_cars1['buying'] = le.fit_transform(df_cars1['buying'])
df_cars1['maint'] = le.fit_transform(df_cars1['maint'])
df_cars1['doors'] = le.fit_transform(df_cars1['doors'])
df_cars1['persons'] = le.fit_transform(df_cars1['persons'])
df_cars1['lug_boot'] = le.fit_transform(df_cars1['lug_boot'])
df_cars1['safety'] = le.fit_transform(df_cars1['safety'])
df_cars1.head()
```

```python
""" Increase also the number of estimators """
X = df_cars1.iloc[:,:-1] 
y = df_cars1['class']
X_train, X_test, y_train, y_test = train_test_split(X, y) 

ada = AdaBoostClassifier(n_estimators=1000, learning_rate=1)

ada.fit(X_train, y_train)
y_pred = ada.predict(X_test)
```

<h4 style="color:#C2F5A6  "> Metrics </h4>
